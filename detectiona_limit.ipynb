# Definition of detection limit
in order to define detection limit the following formula applies:

A = 1.65 * 2 * C/W * sqrt(M * (1/t0 + 1/tb)) 
 

```python
import numpy as np
import pandas as pd
```

```python
t_0 = 3600
t_b = 10800
```

```python
data = pd.DataFrame(columns=['geometry', 'nuclid', 'bg_count_rate', 'C', 'weight'])
data['geometry'] = ['M-1l', 'M-1l', 'M-1l', 'M-1l', 'M-1l', 'M-1l', 'M-05l', 'M-05l', 'M-05l', 'Fl-05l', 'Fl-05l', 'Fl-05l', 'Fl-05l','Fl-01l', 'Fl-01l',  'Fl-01l', 'Fl-01l']
data['nuclid'] = ['I-131', 'Cs-134', 'Cs-137', 'K-40', 'Ra-226', 'Th232', 'Cs-134', 'Cs-137', 'K-40','I-131', 'Cs-134', 'Cs-137', 'K-40', 'I-131', 'Cs-134', 'Cs-137', 'K-40']
data['bg_count_rate'] = [6, 1.5, 2, 1.2, 0.3, 0.15, 1.5, 2, 1.2, 6, 1.5, 2, 1.2, 6, 1.5, 2, 1.2]
data['C'] = [528, 243, 220, 14.5, 54.5, 46, 148, 159, 9.3, 163, 75, 68, 4.54, 67, 31, 28, 1.72]
data['weight'] = [1, 1, 1, 1, 1, 1, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.1, 0.1, 0.1, 0.1]
data['C'] = data['C'] * 0.0001
data['A'] = 1.65 * 2 * 1/data['C'] * np.sqrt(data['bg_count_rate'] * (1/t_0) + (1/t_b))
data
```

```python
detection_limit =  data[['geometry', 'nuclid', 'A']]
dl = pd.pivot_table(detection_limit, 
                    index='geometry',
                    columns='nuclid',
                    values='A'
                    )
dl
```
