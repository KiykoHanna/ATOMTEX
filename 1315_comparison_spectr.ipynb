```python
import os
from turtle import color
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from turtle import color
import seaborn as sns
from sklearn.linear_model import LinearRegression
from pathlib import Path
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
# read the file with database
url =' https://raw.githubusercontent.com/KiykoHanna/ATOMTEX/main/1315_database.csv'
db = pd.read_csv(url)
db.columns = ['name',
              'nuclid',
              'time',
              'serial_number',
              'geometry',
              'weight',
              'date',
              'sp_beta',
              'Energy',
              'sp_gamma',
              'R_cal',
              'comments'
              ]
db.head(2)

```

```python
# convert some columns from string to list
def convert_to_list(a_str):
    return [float(i) for i in a_str[1:-1].split(', ')]

list_of_str = ['Energy', 'R_cal', 'sp_gamma', 'sp_beta']
for i in list_of_str:
    db[i] = db[i].apply(convert_to_list)
```

```python
# create function for select and show

# function for searching spectrs including a_sytring in the name
def select_name(data, a_str, col_name='name'):
    a = data[col_name]
    b = [i if a_str in i else '---' for i in a]
    data = data.loc[data[col_name].isin(b)]
    return data

# function for showing spectrs
def show_comp_spectr(data):
    fig = make_subplots(rows=2, cols=1)
    color_continuous_scale = ["orange", "red", "green", "blue", "purple", 'turquoise', 'violet', 'wheat', 'yellow', 'yellowgreen']
    for j, i in enumerate(data.index):
        fig.add_trace(
            go.Scatter(x=data['Energy'][i],
                        y=data['sp_gamma'][i],
                        mode="markers",
                        marker=dict(size=3, color= color_continuous_scale[j % len(color_continuous_scale)]),
                        name=data['name'][i] +'gamma'),
            row=1, col=1,

        )

    for j, i in enumerate(data.index):
        fig.add_trace(
            go.Scatter(x=data['Energy'][i],
                        y=data['sp_beta'][i],
                        mode="markers",
                        marker=dict(size=3, color= color_continuous_scale[j % len(color_continuous_scale)]),
                        name=data['name'][i] + 'beta'),
            row=2, col=1
        )

    fig.update_layout(height=600, width=800, title_text="Spectrs")
    fig.show()
```

```python
# select file to show

search = 'Мар Г3 10-2.5175'

data = select_name(db, search, col_name='name')
print(data.name)
len(data)
```
