```python
class Device_1125():
  def __init__(self, number, var):
    self.number = number
    self.var = var

  def calc_coeff(self, act, d):
    self.coeff = act / (np.mean(d))
    return self.coeff

  def calc_point_err(self, act, d):
    mn = np.mean(d, axis=1)
    err = np.round(100  * (-act + mn*1.75) / act, 1)
    self.err = err

    print(mn, '\n', mn*1.75, '\n', self.err)



  def set_bg(self, bg_v, bg_p):
    self.bg_v = bg_v
    self.bg_p = bg_p

  def print_dev(self):
    print(self.number, ',  ', self.var, '\n', 'background volume: ', self.bg_v,  '\n', 'coefficient: ', self.coeff, '\n', )
```
